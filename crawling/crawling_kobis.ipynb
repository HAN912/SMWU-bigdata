## |좌석수 변수 크롤링

```python
seat1=pd.read_excel(r"C:\Users\yangj\Desktop\KOBIS_기간별_좌석점유율.xlsx",encoding='cp949')
seat1.columns=['순위','영화명','개봉일','좌석판매율','좌석점유율','좌석수','매출액','누적매출액','관객수','누적관객수']
seat1.drop([0,1,2,3],inplace=True)
seat1_=seat1.drop(['순위','매출액','누적매출액'],axis=1)
seat1_.head()
```

```python
seat2=pd.read_excel(r"C:\Users\yangj\Desktop\KOBIS_기간별_좌석점유율(1).xlsx",encoding='cp949')
seat2.columns=['순위','영화명','개봉일','좌석판매율','좌석점유율','좌석수','매출액','누적매출액','관객수','누적관객수']
seat2.drop([0,1,2,3],inplace=True)
seat2_=seat2.drop(['순위','매출액','누적매출액'],axis=1)
seat2_.head()
```

```python
A=pd.merge(seat1_,seat2_,how='outer',on=['영화명','개봉일'])
A=A.drop(['좌석판매율_x','좌석점유율_x','좌석판매율_y','좌석점유율_y'],axis=1)
A=A.fillna(0)
A['좌석수']=list(A['좌석수_x']+A['좌석수_y'])
A.head()
```

```python
A=A.drop(['좌석수_x','관객수_x','누적관객수_x','좌석수_y','관객수_y','누적관객수_y'],axis=1)
A.head()
```

```python
#좌석수가 다 있음
kk1=pd.merge(main,A,on=['영화명','개봉일'])
len(kk1)
```

```python
kk1
```

## |좌석수 변수가 존재하는 2004년 이후 영화 데이터만 추출
