## 1. 네이버 영화 평점 크롤링

```python
movie =data.영화명.unique()
scorelist = []
i=0
for i in range(len(movie)):
    url = 'https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query=영화 '+str(movie[i])
    response = requests.get(url)
    html = response.text
    bs = BeautifulSoup(html, 'html.parser')
    if(len(bs.find_all('div',{'class' : 'info_group'}))>=3):
        score_str = bs.find_all('div',{'class' : 'info_group'})[2]
        score = score_str.get_text()[4:8]
        scorelist.append(score)
    else:
        score="없음"
        scorelist.append(score)
print(scorelist)
```

```python
#type((scorelist))
data['평점']= score_list
```

## 2. 네이버 뉴스 기사 수 크롤링

```python
data
```

```python
year=data.개봉일
k=data.영화명
h=[]
for i in range(len(k)):
    h=0
    for page in range(1,400,10):
        query="영화 "+str(k[i])
        now = year[i]
        s=(now-relativedelta(months=1)).strftime('%Y.%m.%d')
        e=(now+relativedelta(months=1)).strftime('%Y.%m.%d')
        url="https://search.naver.com/search.naver?&where=news&query="+query+"&sm=tab_pge&sort=0&photo=0&field=0&reporter_article=&pd=3&ds="+s+"&de="+e+"&docid=&nso=so:r,p:from20200615to20200815,a:all&mynews=0&cluster_rank=228&start="+str(page)+"&refresh_start=0"
        # url 주소에 데이터 요청
        response = requests.get(url)

        # 응답 데이터 : HTML 소스 가져오기
        html = response.text

        # BeautifulSoup으로 html소스를 python 객체로 변환
        soup = BeautifulSoup(html, "html.parser")
                    
        title = soup.find("div", {'class' : 'section_head'}).find("span")
        num=int(title.get_text().split("/ ")[1].replace("건","").replace(",",""))
        h=h+num
    news.append(h)
    print(query)
## 잘못한 점 : news 리스트에 174부터 끝까지 데이터를 추가시킴 원래는 새로운 리스트에 넣으려고 했는데
```

```python
print(news)
print(len(news))
```

```python
data['뉴스']=news
```

## 3. 네이버 영화 상영시간 크롤링

```python
movie =data.영화명.unique()
timelist = []
i=0
for i in range(len(movie)):
    url = 'https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query=영화 '+str(movie[i])
    response = requests.get(url)
    html = response.text
    bs = BeautifulSoup(html, 'html.parser')
    if(len(bs.find_all('div',{'class' : 'info_group'}))>=1):
        time_str = bs.find_all('div',{'class' : 'info_group'})[0].get_text()
        try:
            time=int(re.findall('\d+',time_str)[0])
            timelist.append(time)
        except:
            time="없음"
            timelist.append(time)
    else:
        time="없음"
        timelist.append(time)
    if i%50==0:
        print(i)
print(timelist)
```

```python
data['상영시간']=timelist
```

## 4. 네이버 영화 대표 장르 크롤링

```python
movie =data.영화명.unique()
gernelist = []
i=0
for i in range(len(movie)):
    url = 'https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query=영화 '+str(movie[i])
    response = requests.get(url)
    html = response.text
    bs = BeautifulSoup(html, 'html.parser')
    if(len(bs.find_all('div',{'class' : 'info_group'}))>=1):
        gerne_str = bs.find_all('div',{'class' : 'info_group'})[0].get_text()
        try:
                gerne=gerne_str[4:7]
                gernelist.append(gerne)                
        except:
            gerne="없음"
            gernelist.append(gerne)
    else:
        gerne="없음"
        gernelist.append(gerne)
    if i%50==0:
        print(i)
print(gernelist)
```

```python
data['장르2']=gernelist
```
