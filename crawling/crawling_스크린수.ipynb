## Open API를 통해서 영화별 전국 스크린수 크롤링

```python
import numpy as np
import pandas as pd

import urllib.request as ul
import requests
import json
```

```python
data=pd.read_excel("real_final.xlsx")
data
```

```python
import datetime
def date_str(x):
    return x.strftime('%Y%m%d')
data["개봉일"]=data["개봉일"].apply(lambda x:date_str(x))
data
```

```python
key = #발급받은키 값을 입력합니다.

weekGb="0" #개봉주의 주말 스크린수를 확인하기 위해서
```

```python
movie_list = []
for i in range(len(data["영화명"])):
    targetDt=data["개봉일"][i]
    url = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchWeeklyBoxOfficeList.json?" + \
      "key=" + str(key) + \
      "&targetDt=" + str(targetDt)+\
      "&weekGb="+str(weekGb)
    request = ul.Request(url)
    response = ul.urlopen(request)
    rescode = response.getcode()
    if rescode == 200: # 요청이 성공적이라면, read.
        responseData = response.read()
    responseData_kr = json.loads(responseData.decode('UTF-8')) 
    for data_cut in responseData_kr['boxOfficeResult']['weeklyBoxOfficeList']:
        if list(data_cut.values())[5]==data["영화명"][i]:
                value_list=[]
                value_list.append(list(data_cut.values())[5])
                value_list.append(list(data_cut.values())[16])
                value_list.append(list(data_cut.values())[17])
                value_list.append(int(targetDt))
                movie_list.append(value_list)
```

```python
movie_list
```

```python
# key list 생성
key_list = ["영화명","스크린수","상영횟수","개봉날짜"]
```

```python
movie_data=pd.DataFrame(movie_list, columns=key_list)
movie_data
```

```python
movie_data.to_excel('movie_API.xlsx',  index=False)
```

## 전체 데이터에 스크린수 업데이트

```python
import numpy as np
import pandas as pd
import datetime

#개봉날짜를 개봉일로 변수명 변경
movie_data=pd.read_excel("movie_API.xlsx")
movie_data.rename(columns={'개봉날짜':'개봉일'},inplace=True)

#개봉일 int를 날짜로 변경
date=[]
for i in range(len(movie_data)):  
    s = str(movie_data['개봉일'][i])
    s_datetime = datetime.datetime.strptime(s, '%Y%m%d')
    date.append(s_datetime)

movie_data['개봉일']=date

def date_str(x):
    return x.strftime('%Y%m%d')
movie_data["개봉일"]=movie_data["개봉일"].apply(lambda x:date_str(x))

movie_data.head()
```

```python
data=pd.read_excel("real_final.xlsx")
data=data.drop(['Unnamed: 0','Unnamed: 0.1'],axis=1) #Unnamed: 0.1 이 생겨서 또 삭제함
data["개봉일"]=data["개봉일"].apply(lambda x:date_str(x))
data.head()
```

```python
data_subset=data.iloc[:,[0,1,14]]
Merge=pd.merge(data_subset,movie_data,how='outer',on=['영화명','개봉일'])
Merge
```

```python
import math

scr=[]
a=0
for i in range(len(Merge['영화명'])):
    if math.isnan(Merge['스크린수'][i])==False:
        a=Merge['스크린수'][i]
        scr.append(a)
    else :
        a=Merge['전국스크린수'][i]
        scr.append(a)
```

```python
data['new스크린수']=scr
```
