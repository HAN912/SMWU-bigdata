```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import KNNImputer
```

```python
data=pd.read_excel('real_final2.xlsx')
data.head()
```

```python
#data=data.drop(['Unnamed: 0','Unnamed: 0.1'],axis=1) #Unnamed: 0.1 이 생겨서 또 삭제함
```

```python
#변수별 null값 있는지 확인
data.info()
```

```python
#new스크린수에 5개 데이터가 없기 때문에 knnimpute 를 이용하여 채우기
screen_init=pd.DataFrame(data["new스크린수"])
imputer=KNNImputer(n_neighbors=5)
screen=imputer.fit_transform(screen_init)
```

```python
data["new스크린수"]=screen
data.info()
```

```python
#계절데이터가 연속형으로 들어가있기 때문에 범주형으로 바꿔주기
season_init=data["계절"]
season=pd.get_dummies(season_init)
data['겨울']=season.iloc[:,0]
data['봄']=season.iloc[:,1]
data['여름']=season.iloc[:,2]
data['가을']=season.iloc[:,3]
data.head()
```

```python
data.columns
```

```python
input_var=['new배급사', '평점', '뉴스', '경쟁 영화수', '상영시간',
           '한국', '등급.1',  '개봉월', '요일', '주말',
           '애니메이션', '액션', '드라마', '코미디', '범죄', '멜로/로맨스', '스릴러', '공포', '판타지', '미스터리',
           '모험', 'SF', '가족', '뮤지컬', '공연', '전쟁', '다큐멘터리', '사극', '느와르','new스크린수',
           '겨울','봄','여름','가을']
target=['누적관객수']
```

```python
train=data[input_var]
test=data[target]
train.rename(columns={'등급.1':'관람등급'},inplace=True)
train.rename(columns={'new배급사':'영화배급사'},inplace=True)
train.rename(columns={'new스크린수':'스크린수'},inplace=True)
```

```python
train
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, test, test_size = 0.4, random_state= 42)
```

```python
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size = 0.5, random_state= 42)
```

```python
print(X_train.shape)
print(X_val.shape)
print(X_test.shape)
```

```python
#real_final2에서 계절을 범주형변수로 바꾸고 new스크린수 생성한 자료
data.to_excel('real_final3.xlsx')
```

## Naive benchmark

```python
y_naive = np.mean(y_train)
rmse_naive = np.sqrt( np.mean( (y_val - y_naive)**2 ) )
print(rmse_naive)
```

```python
# 누적관갹수에 대하여 로그변환
y_naive = np.mean(np.log1p(y_train))
rmse_naive = np.sqrt( np.mean( (np.log1p(y_val) - y_naive)**2 ) )
print(rmse_naive)
```

```python
# 누적관갹수에 대하여 로그변환(MAE)
mae_naive = np.mean( abs( (np.log1p(y_val) - y_naive) ) )
print(mae_naive)
```
